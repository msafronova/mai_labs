#### Найти ближайшее к заданному значению число в заданном массиве

````python
arr = np.array([1, 2, 3, 4, 5, 6, 7, 10])

def find_nearest(array: np.ndarray, value):
    idx = (np.abs(array - value)).argmin()
    return array[idx]

find_nearest(arr, 8)
````
